```python
import pandas as pd
combined_data = pd.read_csv('./Data/all_players_stats.csv')
```

```python
combined_data.head()
```

```python
sorted_players = combined_data.sort_values('Weighted_Score', ascending=False)
num_players_per_team = 11

man_city_players = sorted_players[sorted_players['Team'] == 'Manchester City'].head(num_players_per_team)
inter_milan_players = sorted_players[sorted_players['Team'] == 'Inter Milan'].head(num_players_per_team)
```

```python
print("Manchester City Team Selection:")
print(man_city_players[['Name', 'Weighted_Score']])
print()
print("Inter Milan Team Selection:")
print(inter_milan_players[['Name', 'Weighted_Score']])
```

```python
Manchester_City_Players_Weight = man_city_players['Weighted_Score'].sum()
Inter_Milan_Players_Weight = inter_milan_players['Weighted_Score'].sum()
```

```python
print("Manchester City Players points: ", Manchester_City_Players_Weight)
print("Inter Milan Players points: ", Inter_Milan_Players_Weight)
```

```python
all_players = pd.concat([man_city_players, inter_milan_players], axis=0)
all_players.sort_values(['Weighted_Score'], ascending=False, inplace=True)
```
